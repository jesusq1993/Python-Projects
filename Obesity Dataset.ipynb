```python
obesity_data_2011 = obesity[obesity['YearStart']==2011]
```

```python
gender_obesity_data_2011 = obesity_data_2011[obesity_data_2011['StratificationCategory1'] == 'Gender']
income_obesity_data_2011 = obesity_data_2011[obesity_data_2011['StratificationCategory1'] == 'Income']
education_obesity_data_2011 = obesity_data_2011[obesity_data_2011['StratificationCategory1'] == 'Education']
age_obesity__data_2011 = obesity_data_2011[obesity_data_2011['StratificationCategory1'] == 'Age (years)']
```

```python
#Graphed the top ten states with the highest obesity rate

#seperated the male data
#sorted the obesity rates from least to greatest
men_obesity_data_2011 = gender_obesity_data_2011[gender_obesity_data_2011['Stratification1'] == 'Male']
men_obesity_data_2011 = men_obesity_data_2011.sort_values('Data_Value', ascending = False)


#saved obesity rate and location into a variable
Location_Desc_2011 = men_obesity_data_2011['LocationDesc'].head(10)
Obesity_rate = men_obesity_data_2011['Data_Value'].head(10)


#values in line graph
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure 
figure(figsize=(17,10))
plt.plot(Location_Desc_2011, Obesity_rate, color = 'blue', label = 'Obesity Rate')
#plt.xticks(rotation = 'vertical')
plt.ylabel('Obesity Rate %')
plt.xlabel('State')
plt.title('Male Obesity Rate for States')
plt.legend(loc = 'upper left')
#plt.style.use('seaborn-white')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
#Graphed the states with the highest obesity rates for females

#seperated the female data
#sorted obesity rates from least to greatest
female_obesity_data_2011 = gender_obesity_data_2011[gender_obesity_data_2011['Stratification1'] == 'Female']
female_obesity_data_2011 = men_obesity_data_2011.sort_values('Data_Value', ascending = False)

#saved top 10 obesity rate and and locations for women in a variable
Location_Desc_2011 = female_obesity_data_2011['LocationDesc'].head(10)
Obesity_rate_women = female_obesity_data_2011['Data_Value'].head(10)

#plot female obesity rate for states
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure 
figure(figsize=(17,10))
plt.plot(Location_Desc_2011, Obesity_rate_women, color = 'blue', label = 'Obesity Rate')
#plt.xticks(rotation = 'vertical')
plt.ylabel('Obesity Rate %')
plt.xlabel('State')
plt.title('Female Obesity Rate for States')
plt.legend(loc = 'upper left')
#plt.style.use('seaborn-white')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
#plotted the obesity rate averages based on education

#save data for individuals with less than highschool education in variable and added it to average list
edu_less_than_highschool_obesity_data_2011 = education_obesity_data_2011[education_obesity_data_2011['Stratification1'] == 'Less than high school']
education_averages = edu_less_than_highschool_obesity_data_2011['Data_Value'].sum()/len(edu_less_than_highschool_obesity_data_2011)

#save data for individuals with some college or technical education in variable and added it to average list
edu_some_obesity_data_2011 = education_obesity_data_2011[education_obesity_data_2011['Stratification1'] == 'Some college or technical school']
temp_education_averages = edu_some_obesity_data_2011['Data_Value'].sum()/len(edu_some_obesity_data_2011)
education_averages = np.append(education_averages, temp_education_averages)

#save data for individuals that are highschool graduates in variable and added it to average list
edu_highschool_obesity_data_2011 = education_obesity_data_2011[education_obesity_data_2011['Stratification1'] == 'High school graduate']
temp_education_averages = edu_highschool_obesity_data_2011['Data_Value'].sum()/len(edu_highschool_obesity_data_2011)
education_averages = np.append(education_averages, temp_education_averages)

#save data for individuals that are college graduate education in variable and added it to average list
edu_college_obesity_data_2011 = education_obesity_data_2011[education_obesity_data_2011['Stratification1'] == 'College graduate']
temp_education_averages = edu_college_obesity_data_2011['Data_Value'].sum()/len(edu_college_obesity_data_2011)
education_averages = np.append(education_averages, temp_education_averages)

#saved list with education labels
ylabels = ['Less Than High School', 'Some college or technical school', 'High school graduate', 'College graduate']

#plotted the obesity rate averages based on education
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure 
figure(figsize=(17,10))
#plt.bar(education_obesity_data_2011['Stratification1'], education_averages, color = 'red',width = .3, label = 'Obesity Rate')
plt.bar(ylabels, education_averages, color = 'blue',width = .3, label = 'Obesity Rate')
#plt.xticks(rotation = 'vertical')
plt.ylabel('Obesity Rate %')
plt.xlabel('Education')
plt.title('Obesity Rate Based on Education')
plt.legend(loc = 'upper left')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
#plotted graph based on age

#data for individuals with age 18 - 24 in variable
age_18_24_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '18 - 24']
age_average = age_18_24_obesity_data_2011['Data_Value'].sum()/len(age_18_24_obesity_data_2011)

#data for individuals with age 25 - 34 in variable
age_25_34_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '25 - 34']
age_average_temp = age_25_34_obesity_data_2011['Data_Value'].sum()/len(age_25_34_obesity_data_2011)
age_average = np.append(age_average, age_average_temp)

#data for individuals with age 35 - 44 in variable
age_35_44_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '35 - 44']
age_average_temp = age_35_44_obesity_data_2011['Data_Value'].sum()/len(age_35_44_obesity_data_2011)
age_average = np.append(age_average, age_average_temp)

#data for individuals with age 45 - 54 in variable
age_45_54_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '45 - 54']
age_average_temp = age_45_54_obesity_data_2011['Data_Value'].sum()/len(age_45_54_obesity_data_2011)
age_average = np.append(age_average, age_average_temp)

#data for individuals with age 55 - 64 in variable
age_55_64_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '55 - 64']
age_average_temp = age_55_64_obesity_data_2011['Data_Value'].sum()/len(age_55_64_obesity_data_2011)
age_average = np.append(age_average, age_average_temp)

#data for individuals with age 65 or older in variable
age_65_older_obesity_data_2011 = age_obesity__data_2011[age_obesity__data_2011['Stratification1'] == '65 or older']
age_average_temp = age_65_older_obesity_data_2011['Data_Value'].sum()/len(age_65_older_obesity_data_2011)
age_average = np.append(age_average, age_average_temp)

#saved labels for age in list
ylabel = age_obesity__data_2011['Stratification1'].unique()


#plotted obesity rate based on age
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure 
figure(figsize=(17,10))
#plt.bar(education_obesity_data_2011['Stratification1'], education_averages, color = 'red',width = .3, label = 'Obesity Rate')
plt.bar(ylabel, age_average, color = 'blue',width = .3, label = 'Obesity Rate')
#plt.xticks(rotation = 'vertical')
plt.ylabel('Obesity Rate %')
plt.xlabel('Age')
plt.title('Obesity Rate Based on Age')
plt.legend(loc = 'upper left')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
X = age_obesity__data_2011['Data_Value'].values
X_new = X[:,np.newaxis]


Y = age_obesity__data_2011['Stratification1']
mapping = {
    '18 - 24' : 1.0,
    '25 - 34' : 2.0,
    '35 - 44' : 3.0,
    '45 - 54' : 4.0,
    '55 - 64' : 5.0,
    '65 or older' :6.0
}

Y = Y.replace(mapping).values
y_new = Y[:,np.newaxis]
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X_new)
y_std = sc_y.fit_transform(y_new)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

# Create linear regression object
regr = linear_model.LinearRegression()



X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state=1)

X_train = X_train[:,np.newaxis]
print(X_train.shape)

X_test = X_test[:,np.newaxis]
print(X_test.shape)

Y_train = Y_train[:,np.newaxis]
print(Y_train.shape)

Y_test = Y_test[:,np.newaxis]
print(Y_test.shape)
```

```python
# Train the model using the training sets
regr.fit(X_std,y_std)
```

```python
#getting the linear scatterplot ready
def lin_regplot(X, y, model):
    plt.scatter(X, y, c = 'steelblue', edgecolor = 'white', s = 70)
    plt.xlabel('Obesity Rate')
    plt.ylabel('Age')
    plt.plot(X, model.predict(X), color = 'red', lw = 2)
    return None
```

```python
#Display linear scatterplot
lin_regplot(X_std, y_std, regr)
plt.show()
```

```python
#regression coefficients
regr.coef_
```

```python
#intercept
regr.intercept_
```

```python
# mean squared error
mean_squared_error(X_std, regr.predict(X_std))
```

```python
#r2 score
r2_score(X_std, regr.predict(X_std))
```
